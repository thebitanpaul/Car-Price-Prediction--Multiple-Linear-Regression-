```python
x_train, x_test, y_train, y_test = train_test_split(df[['enginesize','stroke','horsepower']],
                                                    df['price'],test_size=0.2)

reg=linear_model.LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_predict=reg.predict(x_test)
print("Accuracy = ", round(reg.score(x_test, y_test)*100,2), "%")
sns.regplot(x=y_predict, y=y_test, ci=0);
plt.show()
```
